```python
import sqlite3
import pandas as pd

#Database connection
connection = sqlite3.connect('powerlifting.db')
cursor = connection.cursor()

# Create the table 
cursor.execute('''
CREATE TABLE IF NOT EXISTS powerlifting_table (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT,
    Sex TEXT,
    Event TEXT,
    Equipment TEXT,
    Age INTEGER,
    Division TEXT,
    BodyweightKg REAL,
    WeightClassKg REAL,
    Squat1Kg REAL,
    Squat2Kg REAL,
    Squat3Kg REAL,
    Bench1Kg REAL,
    Bench2Kg REAL,
    Bench3Kg REAL,
    Deadlift1Kg REAL,
    Deadlift2Kg REAL,
    Deadlift3Kg REAL,
    TotalKg REAL,
    Place TEXT,
    Wilks REAL,
    Country TEXT,
    Federation TEXT,
    Date TEXT
)
''')

#Read the CSV file into a DataFrame
newdf = pd.read_csv('firstupdatepowerlifting.csv')

#Clean the DataFrame (Remove rows with NULL in TotalKg)
newdf = newdf[newdf['TotalKg'].notnull()]

#Clear existing data
cursor.execute('DELETE FROM powerlifting_table')

#Insert the cleaned DataFrame into the table
newdf.to_sql('powerlifting_table', connection, if_exists='append', index=False)

#Remove lifters where Age is NULL
cursor.execute('''
DELETE FROM powerlifting_table
WHERE Age IS NULL;
''')

#Remove duplicates from the database
cursor.execute('''
CREATE TABLE IF NOT EXISTS temp_table AS
SELECT DISTINCT * FROM powerlifting_table;
''')

#Drop original table
cursor.execute('DROP TABLE powerlifting_table;')

#Rename the temporary table to the original table name
cursor.execute('ALTER TABLE temp_table RENAME TO powerlifting_table;')

#Execute a query to retrieve data
query = 'SELECT * FROM powerlifting_table'
result = pd.read_sql_query(query, connection)

print(result)

#Close the connection
#connection.close()
```

```python
import sqlite3
import pandas as pd

# Query Function
def run_query(query):
    cursor.execute(query)
    return cursor.fetchall()

# Query to count males and females
query = '''
SELECT 
    COUNT(CASE WHEN Sex = 'M' THEN 1 END) AS male_count,
    COUNT(CASE WHEN Sex = 'F' THEN 1 END) AS female_count
FROM 
    powerlifting_table;
'''

result = run_query(query)

#Display the result
male_count, female_count = result[0]  
print(f'Number of Males: {male_count}')
print(f'Number of Females: {female_count}')

```

```python
# Define the query to count lifters from each country
query = '''
SELECT 
    Country, 
    COUNT(*) AS lifter_count
FROM 
    powerlifting_table
GROUP BY 
    Country
ORDER BY 
    lifter_count DESC;
'''

result = run_query(query)

# Display the result
for country, count in result:
    print(f'Country: {country}, Number of Lifters: {count}')
```

```python
# How many participants are in each age class? 
query = 'SELECT Age FROM powerlifting_table'
df = pd.read_sql_query(query, connection)

# Categorize ages
def categorize_age(age):
    if 14 <= age <= 18:
        return 'Sub-junior (14-18 yrs)'
    elif 19 <= age <= 23:
        return 'Junior (19-23 yrs)'
    elif age >= 40 and age <= 49:
        return 'Masters 1 (40-49 yrs)'
    elif age >= 50 and age <= 59:
        return 'Masters 2 (50-59 yrs)'
    elif age >= 60 and age <= 69:
        return 'Masters 3 (60-69 yrs)'
    elif age >= 70:
        return 'Masters 4 (70+ yrs)'
    else:
        return 'Open (Any Age)'  

#Create a new column for age classes
df['Age Class'] = df['Age'].apply(categorize_age)

#Group by the new age class and count participants
age_class_counts = df.groupby('Age Class').size().reset_index(name='Participants')

#Display the results
print(age_class_counts)
```

```python
# Execute the SQL query
query = '''
SELECT 
    CASE 
        WHEN Age BETWEEN 14 AND 18 THEN 'Sub-junior'
        WHEN Age BETWEEN 19 AND 23 THEN 'Junior'
        WHEN Age BETWEEN 40 AND 49 THEN 'Masters 1'
        WHEN Age BETWEEN 50 AND 59 THEN 'Masters 2'
        WHEN Age BETWEEN 60 AND 69 THEN 'Masters 3'
        WHEN Age >= 70 THEN 'Masters 4'
        ELSE 'Open'
    END AS Age_Class,
    Sex,
    MAX(Squat3Kg) AS Max_Squat,
    MAX(Bench3Kg) AS Max_Bench,
    MAX(Deadlift3Kg) AS Max_Deadlift
FROM 
    powerlifting_table
GROUP BY 
    Age_Class, Sex
ORDER BY 
    Age_Class, Sex;
'''

#Load the results into a DataFrame
results = pd.read_sql_query(query, connection)
print(results)

#Close the connection
connection.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset 
data = pd.read_csv('firstupdatepowerlifting.csv')

# Converting 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# 1. Bar Chart: Number of participants by Country
participants_by_country = data['Country'].value_counts()
#Get the top 10 countries
top_10_countries = participants_by_country.head(10)
#Create the bar chart
plt.figure(figsize=(12, 6))
top_10_countries.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Participants by Country (Top 10)', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of Participants', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 2. Boxplot: Distribution of Wilks scores by Sex
plt.figure(figsize=(8, 6))
data.boxplot(column='Wilks', by='Sex', grid=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Distribution of Wilks Scores by Sex', fontsize=16)
plt.suptitle('')  # Removes the default title
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Wilks Score', fontsize=14)
plt.tight_layout()
plt.show()

# 3. Line Chart: Plot TotalKg or Wilks over Date to observe trends in performance over time
# Aggregate data by date (mean TotalKg and Wilks per date)
trend_data = data.groupby('Date')[['TotalKg', 'Wilks']].mean().dropna()
plt.figure(figsize=(12, 6))
plt.plot(trend_data.index, trend_data['TotalKg'], label='Average TotalKg', marker='o', linestyle='-')
plt.plot(trend_data.index, trend_data['Wilks'], label='Average Wilks', marker='o', linestyle='-')
plt.title('Performance Trends Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Average Performance', fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 4. Histogram: Distribution of Age
plt.figure(figsize=(8, 6))
plt.hist(data['Age'].dropna(), bins=20, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribution of Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import csv
import pandas as pd
import numpy as np

# Open the input CSV file
with open('firstupdatepowerlifting.csv', 'r') as input_file:
    reader = csv.DictReader(input_file)
    
    # Open the output CSV file
    with open('updatedpowerlifting.csv', 'w', newline='') as output_file:
        writer = csv.DictWriter(output_file, fieldnames=reader.fieldnames)
        writer.writeheader()
        
        # Iterate through each row in the input file
        for row in reader:
            # Check if the "WeightClassKg" column contains a string
            if row['WeightClassKg'].isdigit():
                # Check if the "Place" column contains a float
                try:
                    float(row['Place'])
                    
                    # Remove the '-' prefix from the numeric columns
                    row['Squat1Kg'] = float(row['Squat1Kg'].lstrip('-'))
                    row['Squat2Kg'] = float(row['Squat2Kg'].lstrip('-'))
                    row['Squat3Kg'] = float(row['Squat3Kg'].lstrip('-'))
                    row['Bench1Kg'] = float(row['Bench1Kg'].lstrip('-'))
                    row['Bench2Kg'] = float(row['Bench2Kg'].lstrip('-'))
                    row['Bench3Kg'] = float(row['Bench3Kg'].lstrip('-'))
                    row['Deadlift1Kg'] = float(row['Deadlift1Kg'].lstrip('-'))
                    row['Deadlift2Kg'] = float(row['Deadlift2Kg'].lstrip('-'))
                    row['Deadlift3Kg'] = float(row['Deadlift3Kg'].lstrip('-'))
                    
                    writer.writerow(row)
                except ValueError:
                    continue  
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Load the CSV data
df = pd.read_csv('updatedpowerlifting.csv')

# Encode categorical variables
df = pd.get_dummies(df, columns=['Sex', 'Event', 'Equipment', 'Division', 'Country', 'Federation'])

# Scale the numerical features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['BodyweightKg', 'WeightClassKg', 'Squat1Kg', 'Squat2Kg', 'Squat3Kg', 'Bench1Kg', 'Bench2Kg', 'Bench3Kg', 'Deadlift1Kg', 'Deadlift2Kg', 'Deadlift3Kg', 'TotalKg', 'Wilks']])
df[['BodyweightKg', 'WeightClassKg', 'Squat1Kg', 'Squat2Kg', 'Squat3Kg', 'Bench1Kg', 'Bench2Kg', 'Bench3Kg', 'Deadlift1Kg', 'Deadlift2Kg', 'Deadlift3Kg', 'TotalKg', 'Wilks']] = df_scaled

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X = df.drop(['Name', 'Date', 'Place'], axis=1)
X_imputed = imputer.fit_transform(X)

# Split the data into features and target
y = df['Wilks']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate the model
r2 = r2_score(y_test, model.predict(X_test))
print(f"R-squared score: {r2:.2f}")
```
